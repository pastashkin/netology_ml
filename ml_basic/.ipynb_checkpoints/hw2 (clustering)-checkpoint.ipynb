### 3. Кластеризация
Скормим наши комментарии, а точнее их координаты в алгоритм кластеризации и посмотрим, что у нас получится.

```python
from sklearn.cluster import KMeans
```

```python
# Делаю 9 кластеров, т.к. в среди голосов не оказалось данных из Зеленограда.
kmeans = KMeans(n_clusters=9, random_state=0)
```

```python
geo_clustering = geo[['x', 'y']]
kmeans.fit(geo_clustering)
```

```python
geo_clustering = pd.DataFrame(kmeans.cluster_centers_, columns=['x', 'y'])
geo_clustering['coordinates'] = list(zip(geo_clustering.x, geo_clustering.y))
geo_clustering['coordinates'] = geo_clustering['coordinates'].apply(Point)
```

```python
# Переводим результаты в GeoDataFrame
geo_clustering = gpd.GeoDataFrame(geo_clustering, geometry='coordinates')
# Повторяем пляски с проекциями
geo_clustering.crs = {'init' :'epsg:4326'}
geo_clustering = geo_clustering.to_crs({'init': 'epsg:3857'})
```

```python
ax =  geo_data.plot(column='NAME', edgecolor='black', figsize=[15,15])
geo_centroids.plot(ax=ax, color='red')
geo_clustering.plot(ax=ax, color='blue')

#рисуем легенду
red_patch = mpatches.Patch(color='red', label='Real centroids')
blue_patch = mpatches.Patch(color='blue', label='Clustering centroid')
plt.legend(handles=[red_patch, blue_patch])
plt.show()
```

### 4. Анализ плотности распределения комментариев
Попробуем отрисовать плотности распределения положительных и отрицательных комментариев.
При этом количество кластеров для двух групп подберем по-умному.
Позитивные комментарии прогоним по метрике silhouette
Негативные по сумме квадратов расстояний

```python
# Для умного выбора количества кластеров применим GridSearch с функцией silhouette
from sklearn.metrics import silhouette_score
from sklearn.cluster import MiniBatchKMeans
```

```python
geo_good = geo[geo.comment_class == 1]
geo_bad = geo[geo.comment_class == -1]
geo_good = geo_good.drop(columns='comment_class')
geo_bad = geo_bad.drop(columns='comment_class')
```

```python
for n_good in range(2, 21):
    kmeans_good = MiniBatchKMeans(n_clusters=n_good, batch_size=1000)
    cluster_labels = kmeans_good.fit_predict(geo_good)
    silhouette_avg = silhouette_score(geo_good, cluster_labels)
    print('For ', n_good, ' clusters silhouette score is ', round(silhouette_avg, 5))
```

Похоже 18 кластеров - наш вариант. Обучимся.

```python
kmeans_good = KMeans(n_clusters=18)
kmeans_good.fit(geo_good)
```

Переберем негативные комментарии и оценим квадрат расстояния

```python
k_inertia = []
n_bad = range(2,21)
for k in n_bad:
    kmeans_bad = MiniBatchKMeans(n_clusters=k, batch_size=1000)
    clusters_kmeans = kmeans_bad.fit_predict(geo_bad)
    k_inertia.append(kmeans_bad.inertia_)
plt.plot(n_bad, k_inertia)
plt.plot(n_bad, k_inertia ,'ro')

diff = np.diff(k_inertia)
diff_r = diff[1:] / diff[:-1]
bad_opt = n_bad[np.argmin(diff_r)+1]
print('Optimal k is', bad_opt)
```

Для негативных оптимально сделать 14 кластров. Обучимся.

```python
kmeans_bad = KMeans(n_clusters=14)
kmeans_bad.fit(geo_bad)
```

### Бонус:
Сперва выбрал другой подход к решению задачи. Осталось нес колько полезных функций, позволяющих по координатам определить к какому району принадлежит точка.

```python
def point_in_polygon(point_x, point_y, polygon):
    p = Point(point_x, point_y)
    p = gpd.GeoSeries(p)
    p.crs = {'init' :'epsg:4326'}
    p = p.to_crs({'init': 'epsg:3857'})
    try:
        result = p.intersects(polygon)
    except:
        raise Exception("Bad data")
    return result[0]
```

```python
print('ЦАО - ', point_in_polygon(37.555232, 55.712539, geo_data.geometry[0]),'\n'
      'ЗАО - ', point_in_polygon(37.554210, 55.710278, geo_data.geometry[0]))
```

```python
def district(point_x, point_y, geo_data):
    right_district = 'Unknown'
    for distr in geo_data.index:
        if point_in_polygon(point_x, point_y, geo_data.loc[distr][1]):
            right_district = geo_data.loc[distr][0]
            return right_district
    return right_district
```

```python
print('ЦАО - ', district(37.555232, 55.712539, geo_data),'\n'
      'ЗАО - ', district(37.554210, 55.710278, geo_data))
```

```python
geo_clear = geo[['x', 'y']].drop_duplicates()
geo_districts_set = {}
for point in geo_clear.index:
    geo_districts_set[(geo_clear.loc[point].x, geo_clear.loc[point].y)] = district(geo_clear.loc[point].x, geo_clear.loc[point].y, geo_data)
```

```python
geo['okrug'] = geo.apply(lambda x: geo_districts_set[(x['x'], x['y'])], axis=1)
```
