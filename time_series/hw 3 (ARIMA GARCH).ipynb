```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.tsa.api as smt
from statsmodels.tsa.stattools import adfuller
from scipy.stats import boxcox
%matplotlib inline
```

```python
auto = pd.read_csv('/home/data/netology_ml/time_series/data/hw 3 (ARIMA GARCH)/monthly-us-auto-registration-tho.csv', sep=';')
auto.columns = ['Month','Auto_registration']
df = auto.Auto_registration
df = boxcox(df, 0.5)
```

```python
def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
    
def tsplot(y, lags=None, figsize=(14, 12), style='bmh'):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (5, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))
        qq_ax = plt.subplot2grid(layout, (4, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)
        sm.qqplot(y, line='s', ax=qq_ax)
        
        plt.tight_layout()
    return
```

```python
tsplot(df)
```

```python
# функция для отрисовки оригинала и модели
def tssplot(series1, series2, labels=['One', 'Two']):
    plt.figure(figsize=(15,5))
    plt.plot(series1, label=labels[0])
    plt.plot(series2, '--r', label=labels[1])
    plt.legend()
    plt.grid()
    return plt.show()
```

### ARIMA
Подбираем оптимальные значения

```python
def best_ARIMA_params(series, range_pdq={'p':[0, 4], 'd':[0, 4], 'q':[0, 4]}):
    from sklearn.metrics import mean_squared_error
    import numpy as np
    import statsmodels.tsa.api as smt
    
    min_error = np.inf
    best_params = []
    
    try:
        for p in range(range_pdq['p'][0], range_pdq['p'][1]+1):
            for d in range(range_pdq['d'][0], range_pdq['d'][1]+1):
                for q in range(range_pdq['q'][0], range_pdq['q'][1]+1):
 
                    mdl = smt.ARIMA(series, order=(p,d,q)).fit(method='mle')
                    predict = mdl.predict()
                    
                    if d == 0:
                        test = series
                    else:
                        test = np.diff(series, d)

                    err = mean_squared_error(test, predict.tolist())
                    
                    if err < min_error:
                        min_error = err
                        best_params = [p, d, q]
    except:
        pass
    return best_params
```

```python
best_ARIMA_params(df, {'p':[0, 5], 'd':[0, 4], 'q':[0, 18]})
```

```python
mdl = smt.ARIMA(df, order=(0,0,12)).fit(method='mle')
tssplot(df, mdl.predict(end=len(df)+12), ['Original', 'ARIMA'])
```

### GARCH
Так же подбираем оптимальные параметры и строим модель

```python
def best_GARCH_params(series, range_poq={'p':[0, 4], 'o':[0, 4], 'q':[0, 4]}):
    from sklearn.metrics import mean_squared_error
    import numpy as np
    from arch import arch_model
    
    min_error = np.inf
    best_params = []
    
    for p in range(range_poq['p'][0], range_poq['p'][1]+1):
        for o in range(range_poq['o'][0], range_poq['o'][1]+1):
            for q in range(range_poq['q'][0], range_poq['q'][1]+1):
                if p == 0 and o == 0:
                    continue

                mdl = arch_model(series, p=p,o=o,q=q).fit(update_freq=5, disp='off')
                forecasts = mdl.forecast(horizon=len(series), method='simulation', simulations=5)
                sims = forecasts.simulations
                
                predict = []
                for i in range(len(df)):
                    predict.append(sims.values[-1].T[i].mean())
                
                

                err = mean_squared_error(series, predict)

                if err < min_error:
                    min_error = err
                    best_params = [p, o, q]

    return best_params
```

```python
best_GARCH_params(df, {'p':[0, 5], 'o':[0, 5], 'q':[0, 5]})
```

```python
n_sim = 5
garch_mdl = arch_model(df, p=1,o=0,q=1).fit(update_freq=5, disp='off')
forecasts = garch_mdl.forecast(horizon=len(df), method='simulation', simulations=n_sim)
sims = forecasts.simulations
garch_pred = []
for i in range(len(df)):
    garch_pred.append(sims.values[-1].T[i].mean())
```

```python
tssplot(df, garch_pred, ['Original', 'GARCH'])
```

### Линейная авторегрессия по-простому

```python
# Модернизирую функцию с занятия
def prepareData(series, lag_start=5, lag_end=30, test_size=0.15):
    
    data = pd.DataFrame(series.copy(), columns=['Y'])
    
    # считаем индекс в датафрейме, после которого начинается тестовый отрезок
    test_index = int(len(data)*(1-test_size))
    
    # добавляем лаги исходного ряда в качестве признаков
    for i in range(lag_start, lag_end):
        data["lag_{}".format(i)] = data.Y.shift(i)
    for i in range(lag_start, lag_end):
        data["lag_mean_{}".format(i)] = data.Y.shift(i)[-12:].mean()
        
    data = data.dropna()
    data = data.reset_index(drop=True)
     
    # разбиваем весь датасет на тренировочную и тестовую выборку
    X_train = data.loc[:test_index].drop(["Y"], axis=1)
    y_train = data.loc[:test_index]["Y"]
    X_test = data.loc[test_index:].drop(["Y"], axis=1)
    y_test = data.loc[test_index:]["Y"]
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = prepareData(df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
ln_model.fit(X_train, y_train)
```

```python
#Train RMSE
mean_squared_error(y_train, ln_model.predict(X_train))
```

```python
#Test RMSE
mean_squared_error(y_test, ln_model.predict(X_test))
```
